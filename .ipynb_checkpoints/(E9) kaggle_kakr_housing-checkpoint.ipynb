## RMSE

```python
def rmse(y_test, y_pred):
    return np.sqrt(mean_squared_error(np.expm1(y_test), np.expm1(y_pred)))
```

## Score

```python
def get_scores(models, train, y):
        df = {}

        for model in models:
            model_name = model.__class__.__name__
            X_train, X_test, y_train, y_test = train_test_split(train, y, random_state=random_state, test_size=0.2)
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)
            if model_name in df:
                model_name = model_name + str(models.index(model))
            df[model_name] = rmse(y_test, y_pred)
            score_df = pd.DataFrame(df, index=['RMSE']).T.sort_values('RMSE', ascending=False)
            
        return (score_df)
```

## Grid Search

- Grid Search를 이용하여 하이퍼 파라미터를 탐색합니다.
- 탐색할 파라미터로는 n_estimators, max_depth, learning_rate입니다.

```python
def my_GridSearch(model, train, y, param_grid, verbose=2, n_jobs=5):

    # 1. GridSearchCV 모델로 `model`을 초기화합니다.
    grid_model = GridSearchCV(model, param_grid=param_grid, scoring='neg_mean_squared_error', \
                                  cv=5, verbose=verbose, n_jobs=n_jobs)
    
    # 2. 모델 fitting
    grid_model.fit(train, y)
    
    # 3. 결과값 저장
    grid_model.cv_results_
    params = grid_model.cv_results_['params']
    score = grid_model.cv_results_['mean_test_score']
   
    # 4. 데이퍼 프레임 생성 및 정렬  
    results = pd.DataFrame(params)
    results['score'] = score
    
    results['RMSE'] = np.sqrt(-1 * results['score'])
    results.sort_values(by=['RMSE'])

    return results
```

## 회고

- LightGBM 모델은 학습한 모델 중 가장 빠른 속도로 학습을 진행을 할 수 있었다. 
- xgboost 모델의 하이퍼 파라미터를 찾기 위해 GridSearch를 실행하였는데 학습 속도가 다른 모델에 비하여 굉장히 오래 걸렸다. 반복 수행시마다 내부적으로 교차검증을 수행해 최적회된 반복 수행횟수를 가질 수 있어 학습 속도가 오래 걸렸을 것이라 판단된다.
- xgboost 모델에 Learning_rate 여러 개를 추가하여 GridSearch를 진행하려고 하였으나, 학습 도중에 주피터 노트북이 꺼지는 오류가 났었다. 학습할 하이퍼 파라미터가 많아서 메모리가 가득차서 문제인지 원인을 아직 찾지 못했다. Learning_rate도 학습 시에 중요한 하이퍼 파라미터라고 하는데 주피터 노트북이 꺼지는 문제때문에 하나의 조건으로만 학습한 것이 아쉬운 점이다.
