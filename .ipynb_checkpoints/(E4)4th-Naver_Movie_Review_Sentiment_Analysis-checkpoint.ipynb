### 2.1 데이터 정제하기

앞서 출력한 리뷰 텍스트를 보면 데이터가 전혀 가공되지 않았습니다. 자연어 처리를 위해서는 단어를 의미있는 벡터와 연결해야 합니다. 먼저는 이 텍스트를 word_to_index 딕셔너리를 통하여 숫자 인덱스로 변환시킵니다. 또한 다음과 같은 전처리 과정을 거쳐 리뷰의 모든 문장을 단어 단위로 쪼개고 의미있는 단어만 뽑아줍니다.

### 토큰화 

한국어 자연어 처리를 위해서는 KoNLPy라는 파이썬 패키지를 사용합니다. KoNLPy를 통해 Okt(Open Korea Text), 메캅(Mecab)을 불러와 형태소 분석을 합니다.

### 불용어 제거

토큰화 후에 의미없는 불용어를 다음과 같이 직접 정의하여 제거합니다.  

stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']

### 사전 word_to_index 구성

컴퓨터는 텍스트보다는 숫자를 더 잘 처리 할 수 있으므로 토큰화 후에 텍스트를 숫자로 바꿔줍니다.

```python
from konlpy.tag import Mecab
tokenizer = Mecab()
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']

def load_data(train_data, test_data, num_words=10000):
    
    train_data.drop_duplicates(subset=['document'], inplace=True)  # 데이터 중복 제거
    train_data = train_data.dropna(how = 'any')                    # NaN 결측치 제거
    test_data.drop_duplicates(subset=['document'], inplace=True)   # 데이터 중복 제거
    test_data = test_data.dropna(how = 'any')                      # NaN 결측치 제거
    
    X_train = []
    for sentence in train_data['document']:
        temp_X = tokenizer.morphs(sentence) # 토큰화
        temp_X = [word for word in temp_X if not word in stopwords] # 불용어 제거
        X_train.append(temp_X)

    X_test = []
    for sentence in test_data['document']:
        temp_X = tokenizer.morphs(sentence) # 토큰화
        temp_X = [word for word in temp_X if not word in stopwords] # 불용어 제거
        X_test.append(temp_X)

    words = np.concatenate(X_train).tolist()
    counter = Counter(words)
    counter = counter.most_common(10000-4)
    vocab = ['<PAD>', '<BOS>', '<UNK>', '<UNUSED>'] + [key for key, _ in counter]
    word_to_index = {word:index for index, word in enumerate(vocab)}

    def wordlist_to_indexlist(wordlist):
        return [word_to_index[word] if word in word_to_index else word_to_index['<UNK>'] for word in wordlist]

    X_train = list(map(wordlist_to_indexlist, X_train))
    X_test = list(map(wordlist_to_indexlist, X_test))

    return X_train, np.array(list(train_data['label'])), X_test, np.array(list(test_data['label'])), word_to_index

X_train, y_train, X_test, y_test, word_to_index = load_data(train_data, test_data)
```

우리가 가진 텍스트 데이터를 숫자로 바꿔 보려고 하는데, 텍스트를 숫자로 바꾸려면 위의 딕셔너리가 {텍스트:인덱스} 구조여야 합니다. index_to_word를 이용합니다.

```python
index_to_word = {index:word for word, index in word_to_index.items()}
```

텍스트 데이터를 숫자로 바꿔 표현해 봅니다. encode된 벡터를 decode하여 다시 원래 텍스트 데이터로 복구할 수도 있습니다.

```python
# 문장 1개를 활용할 딕셔너리와 함께 주면, 단어 인덱스 리스트 벡터로 변환해 주는 함수입니다. 
# 단, 모든 문장은 <BOS>로 시작하는 것으로 합니다. 
def get_encoded_sentence(sentence, word_to_index):
    return [word_to_index['<BOS>']]+[word_to_index[word] if word in word_to_index else word_to_index['<UNK>'] for word in sentence.split()]

# 여러 개의 문장 리스트를 한꺼번에 단어 인덱스 리스트 벡터로 encode해 주는 함수입니다. 
def get_encoded_sentences(sentences, word_to_index):
    return [get_encoded_sentence(sentence, word_to_index) for sentence in sentences]

# 숫자 벡터로 encode된 문장을 원래대로 decode하는 함수입니다. 
def get_decoded_sentence(encoded_sentence, index_to_word):
    return ' '.join(index_to_word[index] if index in index_to_word else '<UNK>' for index in encoded_sentence[1:])  #[1:]를 통해 <BOS>를 제외

# 여러개의 숫자 벡터로 encode된 문장을 한꺼번에 원래대로 decode하는 함수입니다. 
def get_decoded_sentences(encoded_sentences, index_to_word):
    return [get_decoded_sentence(encoded_sentence, index_to_word) for encoded_sentence in encoded_sentences]
```

### 2.2 모델구성을 위한 데이터 분석 및 가공

각 문장의 길이가 다른데, 길이를 맞춰주지 않으면 컴퓨터 상에서 다른 행렬로 보아 계산이 복잡해집니다. 그러므로 pad_sequences를 통해 데이터셋 상의 문장의 길이를 통일해줍니다. 문장의 최대 길이 maxlen의 값을 설정하는 것이 전체 모델 성능에 영향을 미칩니다. 적절한 값을 찾기 위해 전체 데이터셋의 분포를 확인해줍니다.

### 문장 길이 분포 및 최대 문장 길이 지정

```python
total_data_text = list(X_train) + list(X_test)

# 텍스트데이터 문장길이의 리스트를 생성한 후
num_tokens = [len(tokens) for tokens in total_data_text]
num_tokens = np.array(num_tokens)
# 문장길이의 평균값, 최대값, 표준편차를 계산해 본다. 
print('문장길이 평균 : ', np.mean(num_tokens))
print('문장길이 최대 : ', np.max(num_tokens))
print('문장길이 표준편차 : ', np.std(num_tokens))

# 예를들어, 최대 길이를 (평균 + 2*표준편차)로 한다면,  
max_tokens = np.mean(num_tokens) + 2 * np.std(num_tokens)
maxlen = int(max_tokens)
print('pad_sequences maxlen : ', maxlen)
print('전체 문장의 {}%가 maxlen 설정값 이내에 포함됩니다. '.format(np.sum(num_tokens < max_tokens) / len(num_tokens)))
```

전체 문장의 93%가 문장 길이 41 이내에 포함되므로, 모든 데이터 샘플의 길이를 41로 맞춰줍니다.

### pad_sequences 을 활용한 패딩 추가

RNN은 입력데이터가 순차적으로 처리되어, 가장 마지막 입력이 최종 state 값에 가장 영향을 많이 미치게 됩니다. 그러므로 마지막 입력이 무의미한 padding으로 채워지는 것은 비효율적입니다. 따라서 pre를 사용해 문장 앞에 패딩을 추가합니다.

```python
X_train = keras.preprocessing.sequence.pad_sequences(X_train,
                                                        value=word_to_index["<PAD>"],
                                                        padding='post', 
                                                        maxlen=maxlen)

X_test = keras.preprocessing.sequence.pad_sequences(X_test,
                                                       value=word_to_index["<PAD>"],
                                                       padding='post', 
                                                       maxlen=maxlen)


print(X_train.shape)
```

## 3. 모델 구성 및 훈련

모델은 3가지 이상 다양하게 구성하여 실험해 보세요.

모델 학습 전에 데이터셋을 3등분으로 나누어 줍니다. train_data, validation_data, test_data

데이터셋의 비율은 대체적으로 6 : 2 : 2 를 가장 많이 쓰는데, 이렇게 나누는 방법을 Simple Validation 이라고 합니다.  
- 훈련용 데이터셋 146,182건 중 30,000건을 분리하여 검증셋(validation set)으로 사용하였습니다.   
- 훈련용 데이터셋은 116,182개로 구성됩니다. 

```python
# validation set 30000건 분리
x_val = X_train[0:30000]   
y_val = y_train[0:30000]

# validation set을 제외한 나머지 120000건
partial_x_train = X_train[30000:]  
partial_y_train = y_train[30000:]

print(partial_x_train.shape)
print(partial_y_train.shape)
```

텍스트 데이터를 다루는데 주로 사용되는 모델은 RNN입니다. RNN 모델은 이전 스텝의 텍스트 데이터와 현재 시점의 텍스트 데이터를 동시에 입력을 주어 출력에 영향을 줍니다. 시퀀스 데이터를 다루는데 처리하기 최적인 모델입니다.

### 1. RNN

```python
vocab_size = 10000    # 어휘 사전의 크기입니다(10,000개의 단어)
word_vector_dim = 200  # 워드 벡터의 차원수 (변경가능한 하이퍼파라미터) # 단어 하나를 표현하는 임베딩 벡터의 차원수입니다. 

model_1 = keras.Sequential()
model_1.add(keras.layers.Embedding(vocab_size, word_vector_dim, input_shape=(None,)))
model_1.add(keras.layers.LSTM(128))   # 가장 널리 쓰이는 RNN인 LSTM 레이어를 사용하였습니다. 이때 LSTM state 벡터의 차원수는 8로 하였습니다. (변경가능)
model_1.add(keras.layers.Dense(8, activation='relu'))
model_1.add(keras.layers.Dense(1, activation='sigmoid'))  # 최종 출력은 긍정/부정을 나타내는 1dim 입니다.

model_1.summary()
```

1-D CNN은 문장 전체를 한꺼번에 한 방향으로 길이 7짜리 필터로 스캐닝하면서 7단어 이내에서 발견되는 특징을 추출하여 그것으로 문장을 분류하는 방식으로 사용됩니다. 이 방식도 텍스트를 처리하는 데 RNN 못지않은 효율을 보여줍니다. CNN은 RNN보다 병렬처리가 효율적이기 때문에 학습속도도 훨씬 빠르게 진행된다는 장점이 있습니다.

### 2. 1-D CNN

```python
vocab_size = 10000  # 어휘 사전의 크기입니다(10개의 단어)
word_vector_dim = 200   # 단어 하나를 표현하는 임베딩 벡터의 차원수입니다. 

model_2 = keras.Sequential()
model_2.add(keras.layers.Embedding(vocab_size, word_vector_dim, input_shape=(None,)))
model_2.add(keras.layers.Conv1D(16, 7, activation='relu'))
model_2.add(keras.layers.MaxPooling1D(5))
model_2.add(keras.layers.Conv1D(16, 7, activation='relu'))
model_2.add(keras.layers.GlobalMaxPooling1D())
model_2.add(keras.layers.Dense(8, activation='relu'))
model_2.add(keras.layers.Dense(1, activation='sigmoid'))  # 최종 출력은 긍정/부정을 나타내는 1dim 입니다.

model_2.summary()
```

GlobalMaxPooling1D() 레이어 하나만 사용하는 방법도 생각해 볼 수 있습니다. 이 방식은 전체 문장 중에서 단 하나의 가장 중요한 단어만 피처로 추출하여 그것으로 문장의 긍정/부정을 평가하는 방식이라고 생각할 수 있는데, 의외로 성능이 잘 나올 수도 있습니다.

### 3. GlobalMaxPooling1D

```python
vocab_size = 10000  # 어휘 사전의 크기입니다(10개의 단어)
word_vector_dim = 200   # 단어 하나를 표현하는 임베딩 벡터의 차원수입니다. 

model_3 = keras.Sequential()
model_3.add(keras.layers.Embedding(vocab_size, word_vector_dim, input_shape=(None,)))
model_3.add(keras.layers.GlobalMaxPooling1D())
model_3.add(keras.layers.Dense(8, activation='relu'))
model_3.add(keras.layers.Dense(1, activation='sigmoid'))  # 최종 출력은 긍정/부정을 나타내는 1dim 입니다.

model_3.summary()
```

```python
# model 학습
# 1. RNN
model_1.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# 2. 1-D CNN
model_2.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# 3. GlobalMaxPooling()
model_3.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
              
epochs=5  # 몇 epoch를 훈련하면 좋을지 결과를 보면서 바꾸어 봅시다. 

# 1. RNN
history_1 = model_1.fit(partial_x_train,
                    partial_y_train,
                    epochs=epochs,
                    batch_size=512,
                    validation_data=(x_val, y_val),
                    verbose=1)

# 2. 1-D CNN
history_2 = model_2.fit(partial_x_train,
                    partial_y_train,
                    epochs=epochs,
                    batch_size=512,
                    validation_data=(x_val, y_val),
                    verbose=1)

# 3. GlobalMaxPooling()
history_3 = model_3.fit(partial_x_train,
                    partial_y_train,
                    epochs=epochs,
                    batch_size=512,
                    validation_data=(x_val, y_val),
                    verbose=1)
```

## 4. 모델 평가

학습이 끝난 모델을 테스트셋으로 평가합니다.

```python
results_1 = model_1.evaluate(X_test,  y_test, verbose=2)
results_2 = model_2.evaluate(X_test,  y_test, verbose=2)
results_3 = model_3.evaluate(X_test,  y_test, verbose=2)

print('1. RNN :',results_1)
print('2. 1-D CNN :',results_2)
print('3. GlobalMaxPooling1D :',results_3)
```

RNN, GlobalMaxPooling1D, 1-D CNN 순으로 정확도가 높습니다. 
- 1. RNN : 0.850
- 2. 1-D CNN : 0.839
- 3. GlobalMaxPooling1D : 0.845

각 모델의 정확도가 비슷하지만, 자연어 처리를 할 때에는 시퀀스 데이터를 다루는 것에 최적인 RNN 모델이 학습이 더욱 잘 되는 것을 확인할 수 있습니다.

## 4.1 Loss 그래프 시각화

model.fit() 과정 중의 train/validation loss, accuracy 등이 매 epoch마다 history 변수에 저장되어 있습니다.
이 데이터를 그래프로 그리면, 수행했던 딥러닝 학습이 잘 진행되었는지, 오버피팅 혹은 언더피팅하지 않았는지 성능을 개선할 수 있는 아이디어를 얻는 좋은 지표가 됩니다.

```python
history_dict_1 = history_1.history
history_dict_2 = history_2.history
history_dict_3 = history_3.history

print(history_dict_1.keys()) # epoch에 따른 그래프를 그려볼 수 있는 항목들
```

```python
import matplotlib.pyplot as plt

# 1. RNN 
acc_1 = history_dict_1['accuracy']
val_acc_1 = history_dict_1['val_accuracy']
loss_1 = history_dict_1['loss']
val_loss_1 = history_dict_1['val_loss']

# 2. 1-D CNN
acc_2 = history_dict_2['accuracy']
val_acc_2 = history_dict_2['val_accuracy']
loss_2 = history_dict_2['loss']
val_loss_2 = history_dict_2['val_loss']

# 3. GlobalMaxPooling1D
acc_3 = history_dict_3['accuracy']
val_acc_3 = history_dict_3['val_accuracy']
loss_3 = history_dict_3['loss']
val_loss_3 = history_dict_3['val_loss']


# 1. RNN 
epochs_1 = range(1, len(acc_1) + 1)

# "bo"는 "파란색 점"입니다
plt.plot(epochs_1, loss_1, 'bo', label='Training loss')
# b는 "파란 실선"입니다
plt.plot(epochs_1, val_loss_1, 'b', label='Validation loss')
plt.title('1. Training and validation loss of RNN')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()

# 2. 1-D CNN
epochs_2 = range(1, len(acc_2) + 1)

# "bo"는 "파란색 점"입니다
plt.plot(epochs_2, loss_2, 'bo', label='Training loss')
# b는 "파란 실선"입니다
plt.plot(epochs_2, val_loss_2, 'b', label='Validation loss')
plt.title('2. Training and validation loss of 1-D CNN')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()

# 3. GlobalMaxPooling1D
epochs_3 = range(1, len(acc_3) + 1)

# "bo"는 "파란색 점"입니다
plt.plot(epochs_3, loss_3, 'bo', label='Training loss')
# b는 "파란 실선"입니다
plt.plot(epochs_3, val_loss_3, 'b', label='Validation loss')
plt.title('3. Training and validation loss of GlobalMaxPooling1D')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

각 모델이 공통적으로 epochs 수가 늘어날 수록 training loss가 줄어듭니다. 데이터를 학습시키는 과정이므로 epoch 수가 더 늘어나면 0에 가까이 갈 것입니다.

그러나, validation loss는 모두 줄어들다가 증가하는 모습을 보입니다.
- RNN : epoch 4에서 최소값을 갖고 증가함.
- 1-D CNN : epoch 2에서 최소값을 갖고 증가함.
- GlobalMaxPooling1D : epoch 2에서 최소값을 갖고 증가함.
  
**validation loss가 줄어들다가 증가하는 이유**는 모델이 **과적합(Overfitting)**되기 때문입니다. validation_data가 도입된 이유가 기존에 train_data와 test_data만으로 모델을 학습할 경우에는 학습과정에서 참조하지 않았던 데이터를 예측하지 못하는 단점을 보완하기 위함입니다. 즉, Overfitting을 방지하기 위해 validation loss를 확인하는데, validation loss가 증가하는 시점부터 Overfitting이 발생했다고 판단하고 이에 따라 학습을 중단합니다.

따라서, 각 모델의 validation loss가 늘어나는 시점을 중단 시점으로 보고 epoch 횟수를 조정해야할 것입니다.

- RNN : epoch 5 -> epoch 4
- 1-D CNN : epoch 5 -> epoch 2
- GlobalMaxPooling1D : epoch 5 -> epoch 2

그 이유는 우리의 모델은 네이버 영화리뷰 데이터셋으로 총 20만개의 리뷰 데이터를 갖고 있지만, 구글에서 제공한 Word2Vec의 한국어 모델은 이보다 더 많은 총 1억개의 단어로 구성되어 있기 때문에 **학습 데이터의 양이 차이가 크게 납니다**. 이를 통해 학습 데이터의 양이 많다면 더 좋은 성능을 만들어낼 수 있다는 것을 알 수 있습니다. 또한, 특정 단어를 기준으로 네이버 영화리뷰보다 뉴스 데이터셋이 더 다양한 단어를 학습할 수 있을 것입니다. 데이터를 구하고자하는 학습 환경도 중요할 것입니다.

이전 스텝에서 학습했던 모델의 임베딩 레이어를 Word2Vec의 것으로 다시 학습시켜 보겠습니다.

```python
vocab_size = 10000    # 어휘 사전의 크기입니다(10,000개의 단어)
word_vector_dim = 200  # 워드 벡터의 차원수 (변경가능한 하이퍼파라미터)

embedding_matrix = np.random.rand(vocab_size, word_vector_dim)

# embedding_matrix에 Word2Vec 워드벡터를 단어 하나씩마다 차례차례 카피한다.
for i in range(4,vocab_size):
    if index_to_word[i] in word2vec:
        embedding_matrix[i] = word2vec[index_to_word[i]]
```

## 5.2 모델 구성 및 훈련

### 1. RNN

```python
vocab_size = 10000    # 어휘 사전의 크기입니다(10,000개의 단어)
word_vector_dim = 200  # 워드 벡터의 차원수 (변경가능한 하이퍼파라미터) # 단어 하나를 표현하는 임베딩 벡터의 차원수입니다. 

model_4 = keras.Sequential()
model_4.add(keras.layers.Embedding(vocab_size, word_vector_dim, input_shape=(None,)))
model_4.add(keras.layers.LSTM(128))   # 가장 널리 쓰이는 RNN인 LSTM 레이어를 사용하였습니다. 이때 LSTM state 벡터의 차원수는 8로 하였습니다. (변경가능)
model_4.add(keras.layers.Dense(8, activation='relu'))
model_4.add(keras.layers.Dense(1, activation='sigmoid'))  # 최종 출력은 긍정/부정을 나타내는 1dim 입니다.

model_4.summary()
```

### 2. 1-D CNN

```python
vocab_size = 10000  # 어휘 사전의 크기입니다(10개의 단어)
word_vector_dim = 200   # 단어 하나를 표현하는 임베딩 벡터의 차원수입니다. 

model_5 = keras.Sequential()
model_5.add(keras.layers.Embedding(vocab_size, word_vector_dim, input_shape=(None,)))
model_5.add(keras.layers.Conv1D(16, 7, activation='relu'))
model_5.add(keras.layers.MaxPooling1D(5))
model_5.add(keras.layers.Conv1D(16, 7, activation='relu'))
model_5.add(keras.layers.GlobalMaxPooling1D())
model_5.add(keras.layers.Dense(8, activation='relu'))
model_5.add(keras.layers.Dense(1, activation='sigmoid'))  # 최종 출력은 긍정/부정을 나타내는 1dim 입니다.

model_5.summary()
```

### 3. GlobalMaxPooling1D

```python
vocab_size = 10000  # 어휘 사전의 크기입니다(10개의 단어)
word_vector_dim = 200   # 단어 하나를 표현하는 임베딩 벡터의 차원수입니다. 

model_6 = keras.Sequential()
model_6.add(keras.layers.Embedding(vocab_size, word_vector_dim, input_shape=(None,)))
model_6.add(keras.layers.GlobalMaxPooling1D())
model_6.add(keras.layers.Dense(8, activation='relu'))
model_6.add(keras.layers.Dense(1, activation='sigmoid'))  # 최종 출력은 긍정/부정을 나타내는 1dim 입니다.

model_6.summary()
```

```python
model_4.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model_5.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model_6.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
              
epochs=5  # 몇 epoch를 훈련하면 좋을지 결과를 보면서 바꾸어 봅시다. 

history_4 = model_4.fit(partial_x_train,
                    partial_y_train,
                    epochs=epochs,
                    batch_size=512,
                    validation_data=(x_val, y_val),
                    verbose=1)
history_5 = model_5.fit(partial_x_train,
                    partial_y_train,
                    epochs=epochs,
                    batch_size=512,
                    validation_data=(x_val, y_val),
                    verbose=1)
history_6 = model_6.fit(partial_x_train,
                    partial_y_train,
                    epochs=epochs,
                    batch_size=512,
                    validation_data=(x_val, y_val),
                    verbose=1)
```
